```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

boston = pd.read_csv('/kaggle/input/dataset-to-perform-regression-and-classification/Boston.csv')
boston
```

As can be seen, the dataframe contains *506* rows and *14* columns, with the latter being different features that can affect the price of a house, such as tax rate.

```python
#creating a grid of scatterplots of all pairs of predictors
g = sns.PairGrid(boston)
g.map_upper(plt.scatter, s=3)
g.map_diag(plt.hist)
g.map_lower(plt.scatter)
g.fig.set_size_inches(12, 12)
```
